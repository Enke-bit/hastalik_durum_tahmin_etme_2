```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
from keras.callbacks import EarlyStopping

# Veri setini yükle
data = pd.read_csv(r"C:\Users\realb\Desktop\hava yolu şirketi yolcu tahmin projesi - 2\airline_passenger_satisfaction.csv")  # Veri setinin dosya yolunu belirtin

# Gereksiz sütunları kaldır
data.drop(columns=['Departure and Arrival Time Convenience', 'Ease of Online Booking'], inplace=True)

# Eksik değerleri doldur veya kaldır
data.dropna(inplace=True)

# Kategorik değişkenleri kodla
label_encoders = {}
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class']:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])

# Veri setini özellikler (X) ve etiketler (y) olarak ayır
X = data.drop(columns=['ID', 'Arrival Delay'])
y = data['Arrival Delay']

# Özellikleri ölçeklendir
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data.drop(columns=['ID', 'Arrival Delay', 'Satisfaction']))


# Veri setini eğitim ve test kümelerine ayır
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# RNN modelini oluştur
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

# Modeli derle
model.compile(optimizer='adam', loss='mean_squared_error')

# Erken durdurma
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Modeli eğit
history = model.fit(X_train.reshape((X_train.shape[0], X_train.shape[1], 1)), y_train, 
                    epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Model performansını değerlendir
loss = model.evaluate(X_test.reshape((X_test.shape[0], X_test.shape[1], 1)), y_test)
print("Test Loss:", loss)
```

```python
import matplotlib.pyplot as plt


# Test veri kümesi üzerinde tahminler yapma
y_pred = model.predict(X_test.reshape((X_test.shape[0], X_test.shape[1], 1)))

# Tahminleri ve gerçek değerleri bir veri çerçevesinde saklama
predictions_df = pd.DataFrame({'Gerçek Değer': y_test.values, 'Tahmin Edilen Değer': y_pred.flatten()})

# Tahminleri ve gerçek değerleri görselleştirme
plt.figure(figsize=(10, 6))
plt.plot(predictions_df.index, predictions_df['Gerçek Değer'], marker='o', label='Gerçek Değer')
plt.plot(predictions_df.index, predictions_df['Tahmin Edilen Değer'], marker='x', linestyle='--', label='Tahmin Edilen Değer')
plt.xlabel('Örnek Numarası')
plt.ylabel('Arrival Delay')
plt.title('Gerçek ve Tahmin Edilen Arrival Delay Değerleri')
plt.legend()
plt.grid(True)
plt.show()

# Tahminlerin doğruluğunu değerlendirme
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Ortalama Kare Hatası (MSE):", mse)
print("R-kare değeri:", r2)
```
